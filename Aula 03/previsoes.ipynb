```python
# Score de crédito do cliente
# Importar base de dados
import pandas as pd

tabela = pd.read_csv('clientes.csv')

display(tabela)
```

```python
# Preparar os dados para IA
display(tabela.info())
display(tabela.columns)

from sklearn.preprocessing import LabelEncoder
# Transformar as colunas de texto em números, pois a IA aprende por probabilidades e não entende texto.
codificador = LabelEncoder()

# Transforma todas as colunas de texto em números, exceto a coluna 'score_credito'
for coluna in tabela.columns:
  if tabela[coluna].dtype == 'object' and coluna != 'score_credito':
    tabela[coluna] = codificador.fit_transform(tabela[coluna])

# Validar se todas as colunas estão numéricas
for coluna in tabela.columns:
  if tabela[coluna].dtype == 'object' and coluna != 'score_credito':
    print(f'Coluna {coluna} ainda é do tipo texto.')

display(tabela)

from sklearn.model_selection import train_test_split
# Separar as colunas de entrada (x) e a coluna de saída (y)
x = tabela.drop(['score_credito','id_cliente'], axis=1) # x: Colunas que o modelo vai usar para aprender
y = tabela['score_credito'] # y: Coluna que o modelo vai prever

# Separar os dados em treino e teste. Treino: Modelo aprende. Teste: Modelo avalia o aprendizado.
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
# Criar o modelo de IA -> prever a nota de crédito

# Importar o modelo de IA
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

# Criar os modelos de IA
modelo_arvore = RandomForestClassifier() # Modelo de árvore de decisão
modelo_knn = KNeighborsClassifier() # Modelo do KNN (nearest neighbors - vizinhos mais próximos)

# Treinar o modelo de IA
modelo_arvore.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)

# Se o modelo de chutasse tudo "Standard", qual seria a acurácia?
contagem_score = tabela['score_credito'].value_counts()
print(contagem_score['Standard'] / sum(contagem_score))
```

```python
# Escolher o melhor modelo de IA
# Avaliar o modelo de IA
from sklearn.metrics import accuracy_score

# Calcular as previsões
previsao_arvore = modelo_arvore.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste.to_numpy())

# Comparar as previsões com os valores reais
print(f'Modelo Árvore de Decisão: {accuracy_score(y_teste, previsao_arvore)}')
print(f'Modelo de KNN: {accuracy_score(y_teste, previsao_knn)}')

# Quais as características mais importantes para definir o score de crédito?
colunas = list(x_teste.columns)
importancia = pd.DataFrame(index=colunas, data=modelo_arvore.feature_importances_)
importancia = importancia * 100
print(importancia.sort_values(by=0, ascending=False))
```
